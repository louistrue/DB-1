# Teil 1: Algorithmen

Ein Algorithmus ist eine **eindeutige Handlungsvorschrift** zur L√∂sung eines Problems, bestehend aus endlich vielen, genau definierten Einzelschritten.

Das Grundprinzip ist immer gleich:

```
Eingabe ‚Üí Verarbeitung nach Regeln ‚Üí Ausgabe
```

Das kennt ihr schon von der von-Neumann-Architektur (Woche 1) und von eurem ersten Python-Programm.

### Beispiel: Algorithmus in Worten vs. Python

**Problem:** Pr√ºfe, ob eine Betonplatte die Mindestdicke von 20 cm einh√§lt.

**Algorithmus in Worten:**
1. Nimm die Dicke der Platte entgegen (Eingabe)
2. Vergleiche die Dicke mit 20 cm (Verarbeitung)
3. Wenn Dicke >= 20 cm: Meldung "OK" (Ausgabe)
4. Sonst: Meldung "Zu d√ºnn" (Ausgabe)

Jetzt dasselbe in Python:

````python
# Algorithmus: Mindestdicke Betonplatte pr√ºfen

dicke_cm = 18  # Eingabe: Dicke in cm

# Verarbeitung nach Regeln
if dicke_cm >= 20:
    ergebnis = "OK: Mindestdicke eingehalten"
else:
    ergebnis = f"Zu d√ºnn! {dicke_cm} cm < 20 cm Minimum"

# Ausgabe
print(ergebnis)
````

Beachtet die Struktur: **Eingabe** (Variable `dicke_cm`), **Verarbeitung** (`if/else`), **Ausgabe** (`print`). Das ist bei jedem Algorithmus gleich.

### Algorithmus vs. Programm

Ein **Algorithmus** beschreibt abstrakt, wie ein Problem zu l√∂sen ist. Ein **Programm** ist die konkrete Umsetzung in einer Sprache.

Derselbe Algorithmus ("Vergleiche zwei Zahlen") sieht in verschiedenen Sprachen unterschiedlich aus, die Logik bleibt aber identisch:

| Sprache | Code |
|---------|------|
| Deutsch | Wenn A gr√∂sser als B, dann gib A aus |
| Python  | `if a > b: print(a)` |
| BASIC   | `IF A > B THEN PRINT A` |

Ihr seht: Die **Logik** ist √ºberall gleich, nur die **Syntax** (Schreibweise) unterscheidet sich.

Hier ein vereinfachtes Beispiel: Der Ablauf einer St√ºtzenbemessung als Algorithmus.

````python
# Vereinfachter Bemessungsablauf als Algorithmus

# --- Eingaben (kommen normalerweise von der Architektin) ---
geschoss = "EG"
last_kn = 450        # Normalkraft in kN
breite_cm = 30       # St√ºtzenbreite in cm
hoehe_cm = 30        # St√ºtzenh√∂he in cm

# --- Verarbeitung ---
flaeche_cm2 = breite_cm * hoehe_cm
flaeche_m2 = flaeche_cm2 / 10000
spannung_mpa = (last_kn / 1000) / flaeche_m2   # MN / m¬≤ = MPa

# Vereinfachte Pr√ºfung: Betondruckfestigkeit C25/30 = 25 MPa
# Mit Sicherheitsfaktor: zul√§ssig ca. 14 MPa
zulaessig_mpa = 14.0

# --- Ausgabe mit Bewertung ---
print(f"St√ºtze {geschoss}: {breite_cm}x{hoehe_cm} cm")
print(f"Normalkraft: {last_kn} kN")
print(f"Spannung: {spannung_mpa:.1f} MPa")
print(f"Zul√§ssig: {zulaessig_mpa} MPa")
print()

if spannung_mpa <= zulaessig_mpa:
    print("‚úÖ Nachweis erf√ºllt")
else:
    print("‚ùå Nachweis NICHT erf√ºllt")
    print(f"   √úberschreitung: {spannung_mpa - zulaessig_mpa:.1f} MPa")
````

In der Vorlesung habt ihr AND, OR und NOT kennengelernt. In Python heissen diese Operatoren `and`, `or` und `not` (kleingeschrieben).

| Logik | Python | Wahr wenn... |
|-------|--------|--------------|
| AND (‚àß) | `and` | **beide** Bedingungen wahr |
| OR (‚à®) | `or` | **mindestens eine** Bedingung wahr |
| NOT (¬¨) | `not` | Bedingung **umgekehrt** wird |

````python
# Logik im Bauingenieurwesen

tragfaehigkeit_ok = True
gebrauchstauglichkeit_ok = False

# AND: Beide m√ºssen erf√ºllt sein
nachweis_erfuellt = tragfaehigkeit_ok and gebrauchstauglichkeit_ok
print(f"Tragf√§higkeit OK: {tragfaehigkeit_ok}")
print(f"Gebrauchstauglichkeit OK: {gebrauchstauglichkeit_ok}")
print(f"Nachweis erf√ºllt (AND): {nachweis_erfuellt}")
print()

# OR: Mindestens eines muss zutreffen
riss_vorhanden = True
durchbiegung_zu_gross = False
sanierung_noetig = riss_vorhanden or durchbiegung_zu_gross
print(f"Riss vorhanden: {riss_vorhanden}")
print(f"Durchbiegung zu gross: {durchbiegung_zu_gross}")
print(f"Sanierung n√∂tig (OR): {sanierung_noetig}")
print()

# NOT: Umkehrung
baubewilligung = False
baustopp = not baubewilligung
print(f"Baubewilligung erteilt: {baubewilligung}")
print(f"Baustopp (NOT): {baustopp}")
````

### Wahrheitstabellen mit Python

Statt die Tabellen von Hand auszuf√ºllen, lassen wir Python die Arbeit machen. So seht ihr gleichzeitig, wie Schleifen funktionieren.

````python
# Wahrheitstabelle AND
print("A | B | A AND B")
print("--+---+--------")
for a in [0, 1]:
    for b in [0, 1]:
        ergebnis = a and b
        print(f"{a} | {b} |   {ergebnis}")
````

````python
# Wahrheitstabelle OR
print("A | B | A OR B")
print("--+---+-------")
for a in [0, 1]:
    for b in [0, 1]:
        ergebnis = a or b
        print(f"{a} | {b} |   {int(bool(ergebnis))}")
````

````python
# Wahrheitstabelle NOT
print("A | NOT A")
print("--+------")
for a in [0, 1]:
    ergebnis = int(not a)
    print(f"{a} |   {ergebnis}")
````

# Teil 3: Bin√§rsystem

Ein Computer kennt nur zwei Zust√§nde: **0** und **1**. Aus diesen Bits wird alles aufgebaut: Zahlen, Text, Bilder, BIM-Modelle.

### Bin√§r zu Dezimal

Jede Stelle hat einen Wert (von rechts: 1, 2, 4, 8, 16, 32, 64, 128). Man addiert nur dort, wo eine 1 steht.

````python
# Bin√§r zu Dezimal: Schritt f√ºr Schritt

binaer = "00010101"

print(f"Bin√§rzahl: {binaer}")
print()
print("Position:  7  6  5  4  3  2  1  0")
print(f"Bit:       {'  '.join(binaer)}")
print(f"Wert:    128 64 32 16  8  4  2  1")
print()

# Berechnung
summe = 0
for i, bit in enumerate(binaer):
    position = len(binaer) - 1 - i
    wert = int(bit) * (2 ** position)
    if int(bit) == 1:
        print(f"  Position {position}: 1 √ó 2^{position} = {wert}")
    summe += wert

print(f"\nSumme: {summe}")
print(f"Kontrolle mit Python: int('{binaer}', 2) = {int(binaer, 2)}")
````

### Dezimal zu Bin√§r

Methode: Fortlaufend durch 2 teilen, Rest notieren, von unten nach oben lesen.

````python
# Dezimal zu Bin√§r: Schritt f√ºr Schritt

zahl = 25

print(f"Dezimalzahl: {zahl}")
print()
print("Division  | Quotient | Rest")
print("----------+----------+-----")

reste = []
aktuell = zahl
while aktuell > 0:
    quotient = aktuell // 2
    rest = aktuell % 2
    print(f"{aktuell:>5} √∑ 2  |    {quotient:>3}   |  {rest}")
    reste.append(str(rest))
    aktuell = quotient

# Von unten nach oben lesen (umkehren)
binaer = ''.join(reversed(reste))
# Auf 8 Stellen auff√ºllen
binaer_8 = binaer.zfill(8)

print(f"\nErgebnis (von unten nach oben): {binaer}")
print(f"Auf 8 Bit aufgef√ºllt: {binaer_8}")
print(f"Kontrolle mit Python: bin({zahl}) = {bin(zahl)}")
````

# Teil 4: Daten, Information, Wissen

Die **Datenpyramide** zeigt, wie aus rohen Zeichen √ºber mehrere Stufen Kompetenz entsteht:

```
  ‚ñ≥ Kompetenz    ‚Üí Erfahrungsbasierte Bewertung
 ‚ñ≥‚ñ≥ Wissen       ‚Üí Vernetzte Informationen
 ‚ñ≥‚ñ≥‚ñ≥ Information  ‚Üí Daten + Kontext
‚ñ≥‚ñ≥‚ñ≥‚ñ≥ Daten        ‚Üí Zeichen + Syntax
‚ñ≥‚ñ≥‚ñ≥‚ñ≥‚ñ≥ Zeichen     ‚Üí Einzelne Symbole
```

Im folgenden Beispiel seht ihr, wie Python-Variablen genau diese Stufen abbilden.

````python
# Die Datenpyramide in Python

# Stufe 1: Zeichen - einzelne Symbole, keine Bedeutung
zeichen = ['6', '.', '5', '0']
print(f"Zeichen:     {zeichen}")
print(f"  ‚Üí Einzelne Symbole, keine Bedeutung")
print()

# Stufe 2: Daten - Zeichen zusammengesetzt nach Syntax
datum = 6.50
print(f"Daten:       {datum}")
print(f"  ‚Üí Eine Zahl, aber was bedeutet sie? Preis? L√§nge? Gewicht?")
print()

# Stufe 3: Information - Daten + Kontext
info = {"Eigenschaft": "Geschossfl√§che", "Wert": 6.50, "Einheit": "m¬≤"}
print(f"Information: {info['Eigenschaft']} = {info['Wert']} {info['Einheit']}")
print(f"  ‚Üí Jetzt wissen wir, was die Zahl bedeutet")
print()

# Stufe 4: Wissen - Information + Einordnung
wissen = {
    "Eigenschaft": "Geschossfl√§che",
    "Wert": 6.50,
    "Einheit": "m¬≤",
    "Norm": "SIA 416",
    "Kategorie": "Hauptnutzfl√§che"
}
print(f"Wissen:      {wissen['Eigenschaft']} nach {wissen['Norm']} = {wissen['Wert']} {wissen['Einheit']}")
print(f"             Kategorie: {wissen['Kategorie']}")
print(f"  ‚Üí Eingeordnet in ein Regelwerk")
print()

# Stufe 5: Kompetenz - Erfahrungsbasierte Bewertung
print(f"Kompetenz:   'Eine Geschossfl√§che von {wissen['Wert']} {wissen['Einheit']}")
print(f"             f√ºr ein ganzes Geb√§ude? Das kann nicht stimmen.'")
print(f"  ‚Üí Nur mit Erfahrung erkennbar, dass dieser Wert unrealistisch ist")
````
