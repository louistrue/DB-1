## Schritt 2: Variablen und Berechnung

Zuse baute den Z3, um Statik-Berechnungen zu automatisieren.
Machen wir genau das: Fl√§chentr√§gheitsmoment eines Holzbalkens (Rechteckquerschnitt).

Die Formel:

$$I_y = \frac{b \cdot h^3}{12}$$

```python
# Querschnitt eines Holzbalkens (Nadelholz C24)

b = 12   # Breite in cm
h = 24   # H√∂he in cm

# Fl√§chentr√§gheitsmoment
I_y = b * h**3 / 12

print(f"Querschnitt: {b} x {h} cm")
print(f"Fl√§chentr√§gheitsmoment I_y = {I_y:,.0f} cm‚Å¥")
```

## Schritt 3: Schleifen (systematisch wiederholen)

Statt eine Berechnung 7x von Hand zu machen, lassen wir den Computer arbeiten.

Aufgabe: F√ºr einen Holzbalken mit fester Breite b = 12 cm, berechne I_y f√ºr verschiedene H√∂hen.

```python
b = 12  # Breite in cm (fix)

print(f"Breite b = {b} cm\n")
print("  h [cm]  |  I_y [cm‚Å¥]")
print("-" * 28)

for h in [16, 20, 24, 28, 32, 36, 40]:
    I_y = b * h**3 / 12
    print(f"    {h:>2}    |   {I_y:>9,.0f}")
```

Beachtet: Wenn sich die H√∂he verdoppelt (16 ‚Üí 32 cm), verachtfacht sich I_y.

Das liegt am $h^3$ in der Formel. Die H√∂he hat einen enormen Einfluss auf die Steifigkeit.

## Schritt 4: Bedingungen (Regeln formulieren)

Der Sprung von ¬´Maschine die rechnet¬ª zu ¬´Maschine die Anweisungen befolgt¬ª.

Wir erg√§nzen eine Plausibilit√§tspr√ºfung:
- I_y < 10'000 cm‚Å¥ ‚Üí zu gering f√ºr eine Deckenspannweite von 5 m
- I_y > 50'000 cm‚Å¥ ‚Üí √ºberdimensioniert f√ºr diese Spannweite
- Dazwischen ‚Üí im Zielbereich

```python
b = 12  # Breite in cm

# Grenzwerte f√ºr Plausibilit√§tspr√ºfung
I_min = 10_000   # cm‚Å¥ (Mindeststeifigkeit)
I_max = 50_000   # cm‚Å¥ (wirtschaftliche Obergrenze)

print(f"Pr√ºfung f√ºr b = {b} cm, Zielbereich I_y: {I_min:,} bis {I_max:,} cm‚Å¥\n")
print("  h [cm]  |   I_y [cm‚Å¥]  |  Bewertung")
print("-" * 50)

for h in [16, 20, 24, 28, 32, 36, 40]:
    I_y = b * h**3 / 12

    if I_y < I_min:
        bewertung = "‚ùå zu gering"
    elif I_y > I_max:
        bewertung = "‚ö†Ô∏è  √ºberdimensioniert"
    else:
        bewertung = "‚úÖ im Zielbereich"

    print(f"    {h:>2}    |    {I_y:>9,.0f}  |  {bewertung}")
```

**Das ist algorithmisches Denken:**
1. Problem definieren (welche Balkenh√∂he ist sinnvoll?)
2. Regeln formulieren (Grenzwerte setzen)
3. Systematisch pr√ºfen lassen
