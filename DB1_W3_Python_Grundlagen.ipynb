# Teil 2: Rechnen und Text

## 2.1 Rechenoperatoren

Python ist ein Taschenrechner – nur mächtiger.

````python
a = 10
b = 3

print("Addition:      ", a + b)      # 13
print("Subtraktion:   ", a - b)      # 7
print("Multiplikation:", a * b)      # 30
print("Division:      ", a / b)      # 3.333...
print("Ganzzahldiv.:  ", a // b)     # 3
print("Rest (Modulo): ", a % b)      # 1
print("Potenz:        ", a ** 2)     # 100
````

| Operator | Bedeutung | Beispiel |
|----------|-----------|----------|
| `+` | Addition | `5 + 3` → `8` |
| `-` | Subtraktion | `5 - 3` → `2` |
| `*` | Multiplikation | `5 * 3` → `15` |
| `/` | Division | `5 / 3` → `1.667` |
| `//` | Ganzzahldivision | `5 // 3` → `1` |
| `%` | Rest (Modulo) | `5 % 3` → `2` |
| `**` | Potenz | `5 ** 3` → `125` |

### Praxisbeispiel: Flächenträgheitsmoment

Das Flächenträgheitsmoment eines Rechteckquerschnitts:

$$I_y = \frac{b \cdot h^3}{12}$$

````python
# Flaechentraegheitsmoment Rechteckquerschnitt

b = 12   # Breite in cm
h = 24   # Hoehe in cm

I_y = b * h**3 / 12

print("Querschnitt:", b, "x", h, "cm")
print("I_y =", I_y, "cm⁴")
````

## 2.2 f-Strings – Text und Variablen kombinieren

Mit **f-Strings** könnt ihr Variablen direkt in Text einbauen. Setzt ein `f` vor die Anführungszeichen und schreibt Variablen in geschweifte Klammern `{}`.

````python
b = 12
h = 24
I_y = b * h**3 / 12

# Ohne f-String (umständlich)
print("I_y = " + str(I_y) + " cm⁴")

# Mit f-String (elegant)
print(f"Querschnitt: {b} x {h} cm")
print(f"I_y = {I_y:,.0f} cm⁴")       # :,.0f = Tausendertrennzeichen, 0 Dezimalstellen
````

**Formatierung in f-Strings:**

| Format | Bedeutung | Beispiel |
|--------|-----------|----------|
| `{x}` | Wert direkt | `12` |
| `{x:.2f}` | 2 Dezimalstellen | `12.00` |
| `{x:,.0f}` | Tausendertrenner, 0 Dezimalen | `13,824` |
| `{x:.1%}` | Prozent mit 1 Dezimale | `75.0%` |

# Teil 3: Entscheidungen und Wiederholungen

## 3.1 Vergleichsoperatoren

Bevor wir Entscheidungen treffen, müssen wir Werte vergleichen können.

````python
x = 25
y = 30

print(f"{x} == {y} : {x == y}")    # gleich?
print(f"{x} != {y} : {x != y}")    # ungleich?
print(f"{x} <  {y} : {x < y}")     # kleiner?
print(f"{x} >  {y} : {x > y}")     # groesser?
print(f"{x} <= {y} : {x <= y}")    # kleiner oder gleich?
print(f"{x} >= {y} : {x >= y}")    # groesser oder gleich?
````

| Operator | Bedeutung | Beispiel |
|----------|-----------|----------|
| `==` | gleich | `5 == 5` → `True` |
| `!=` | ungleich | `5 != 3` → `True` |
| `<` | kleiner als | `3 < 5` → `True` |
| `>` | grösser als | `5 > 3` → `True` |
| `<=` | kleiner oder gleich | `5 <= 5` → `True` |
| `>=` | grösser oder gleich | `5 >= 6` → `False` |

**Achtung:** `=` ist Zuweisung, `==` ist Vergleich!

## 3.2 `if / elif / else` – Entscheidungen treffen

Mit Bedingungen reagiert euer Programm auf verschiedene Situationen.

````python
# Plausibilitaetspruefung: Betonplattendicke

dicke_cm = 18
mindest_dicke = 20

if dicke_cm >= mindest_dicke:
    print(f"✅ {dicke_cm} cm: Mindestdicke eingehalten")
else:
    print(f"❌ {dicke_cm} cm: Zu dünn! Minimum ist {mindest_dicke} cm")
````

**Wichtig:** Der Code-Block nach `if`, `elif` oder `else` wird **eingerückt** (4 Leerzeichen). Die Einrückung zeigt Python, welcher Code zur Bedingung gehört.

````python
# Mehrere Bedingungen mit elif

raumhoehe = 2.35  # Testet auch mit 2.70 und 2.10
nutzung = "Wohnraum"

# Mindesthoehen nach Nutzung
if nutzung == "Wohnraum":
    mindest = 2.40
elif nutzung == "Büro":
    mindest = 2.50
elif nutzung == "Lager":
    mindest = 2.20
else:
    mindest = 2.40
    print(f"⚠️ Nutzung '{nutzung}' unbekannt, verwende Standard 2.40 m")

if raumhoehe >= mindest:
    print(f"✅ {nutzung}: {raumhoehe} m >= {mindest} m")
else:
    differenz = mindest - raumhoehe
    print(f"❌ {nutzung}: {raumhoehe} m < {mindest} m (fehlen {differenz:.2f} m)")
````

**Aufbau:**
```python
if bedingung_1:
    # wird ausgeführt wenn bedingung_1 True ist
elif bedingung_2:
    # wird ausgeführt wenn bedingung_2 True ist
else:
    # wird ausgeführt wenn keine Bedingung True ist
```
- `elif` und `else` sind **optional**
- Es können beliebig viele `elif`-Blöcke verwendet werden

# Teil 4: Struktur

## 4.1 Listen – Mehrere Werte speichern

Eine Liste speichert **mehrere Werte in einer Variable**. Wie ein Ordner mit nummerierten Seiten.

````python
# Liste erstellen
lasten = [120, 85, 95, 200, 150]   # kN

# Einzelne Elemente (Index startet bei 0!)
print("Erstes Element: ", lasten[0])    # 120
print("Zweites Element:", lasten[1])    # 85
print("Letztes Element:", lasten[-1])   # 150

# Anzahl Elemente
print("Anzahl:", len(lasten))

# Element hinzufügen
lasten.append(175)
print("Nach append:", lasten)
````

| Operation | Code | Ergebnis |
|-----------|------|----------|
| Erstellen | `[1, 2, 3]` | Liste mit 3 Elementen |
| Zugriff | `liste[0]` | Erstes Element |
| Letztes | `liste[-1]` | Letztes Element |
| Länge | `len(liste)` | Anzahl Elemente |
| Hinzufügen | `liste.append(x)` | Element anfügen |
| Summe | `sum(liste)` | Summe aller Elemente |
| Minimum | `min(liste)` | Kleinstes Element |
| Maximum | `max(liste)` | Grösstes Element |

````python
# Nuetzliche Funktionen für Listen

lasten = [120, 85, 95, 200, 150]

print(f"Lasten:  {lasten} kN")
print(f"Summe:   {sum(lasten)} kN")
print(f"Minimum: {min(lasten)} kN")
print(f"Maximum: {max(lasten)} kN")
print(f"Mittel:  {sum(lasten) / len(lasten):.1f} kN")
````

## 4.2 Funktionen – Code wiederverwenden

Eine Funktion ist ein **benannter Code-Block**, den ihr beliebig oft aufrufen könnt. Wie eine Formel, die ihr einmal definiert und dann immer wieder anwendet.

````python
# Funktion definieren
def flaechentraegheitsmoment(b, h):
    """Berechnet I_y eines Rechteckquerschnitts in cm⁴."""
    return b * h**3 / 12

# Funktion aufrufen
ergebnis = flaechentraegheitsmoment(12, 24)
print(f"I_y = {ergebnis:,.0f} cm⁴")

# Mehrfach verwenden
print(f"I_y(10, 20) = {flaechentraegheitsmoment(10, 20):,.0f} cm⁴")
print(f"I_y(16, 32) = {flaechentraegheitsmoment(16, 32):,.0f} cm⁴")
````

**Aufbau einer Funktion:**
```python
def funktionsname(parameter1, parameter2):
    """Kurze Beschreibung (optional)."""
    # Berechnung
    ergebnis = ...
    return ergebnis
```
- `def` leitet die Definition ein
- **Parameter** sind die Eingabewerte (in Klammern)
- `return` gibt das Ergebnis zurück

````python
# Funktion mit Bedingung

def pruefe_dicke(dicke_cm, mindest_cm=20):
    """Prüft ob eine Betonplatte die Mindestdicke einhält."""
    if dicke_cm >= mindest_cm:
        return f"✅ {dicke_cm} cm: OK"
    else:
        return f"❌ {dicke_cm} cm: Zu dünn (min. {mindest_cm} cm)"

# Verschiedene Platten prüfen
for dicke in [15, 18, 20, 22, 25]:
    print(pruefe_dicke(dicke))
````
